```python
# Defining the elements to be retained 
retain_elements = ['id', 'favorite_count', 'retweet_count']

# Defining the required Dictionary
tweet_list = []
with open('tweet-json.txt', 'r') as r: 
    json_tweet = r.readline()
    while json_tweet:
        dictionnary = json.loads(json_tweet)
        tweet_row = dict((i, dictionnary[i]) for i in retain_elements)
        tweet_list.append(tweet_row)
        json_tweet = r.readline()
        
            
# create DataFrame    
df_tweet = pd.DataFrame.from_dict(tweet_list)

# confirm
df_tweet.info()
```

### Issue #2: Tidiness

#### twitter_archive_enhanced table:
dog 'stage' ('doggo', 'floofer', 'pupper','puppo')are in separate columns.

#### Define:
Melt columns 'doggo', 'floofer', 'pupper','puppo' into dog_stage column Create one column for dog 'stage' by concat function. For entries with no dog 'stage', replace with NaN  using np.nan function

#### Code

```python
archive_clean['doggo']=archive_clean['doggo'].replace('None','')
archive_clean['floofer']=archive_clean['floofer'].replace('None','')
archive_clean['pupper']=archive_clean['pupper'].replace('None','')
archive_clean['puppo']=archive_clean['puppo'].replace('None','')
```

```python
archive_clean['dog_stage']=archive_clean['doggo'].str.cat(archive_clean['floofer']).str.cat(archive_clean['pupper']).str.cat(archive_clean['puppo'])
archive_clean=archive_clean.drop(['doggo','floofer','pupper','puppo'], axis =1)
archive_clean['dog_stage']=archive_clean['dog_stage'].replace('', np.nan)
```

### Visualization

#### What is the most common dog_stage?

```python
df.info()
```

```python
x=df.dog_stage.value_counts()
x
```

```python
plt.subplots(figsize=(10, 8))
plt.bar(x.index,x)
plt.xlabel("Dog stage",fontsize = 15);
plt.ylabel("Frequency",fontsize = 15);
plt.title("Dog stage popularity",fontsize = 18);
```

#### What is the top 5 breed of dog by total retweet count?

```python
fav_retweet=df[df.prediction_1 != 'none']
fav_retweet=fav_retweet.groupby(['prediction_1'])
fav_retweet=fav_retweet['retweet_count'].sum()
fav_retweet=fav_retweet.sort_values(ascending =False)
fav_retweet=fav_retweet.head(5)
```

```python
plt.barh(fav_retweet.index, fav_retweet,color='gold')

plt.xlabel('Total Retweet Count', fontsize = 14)
plt.ylabel('Dog Breed', fontsize = 14)
plt.title('Top 5 Dog Breeds by Total Reweet Count', fontsize = 16)
plt.show();
```

#### Is the a relationship between  favorite count and retweet_count?

```python
sb.regplot(x=df.retweet_count, y=df.favorite_count)
plt.xlabel('Number of retweet_count', fontsize = 14)
plt.ylabel('Number of favorite_count', fontsize = 14)
plt.title('Scatterplot for No. of favorite_count vs No. of retweet_count', fontsize = 14)
```

#### What is the distribution of both rating_numerator and rating_denominator?

```python
plt.hist(df['rating_numerator'].value_counts(), bins= 15, color='green', rwidth=20,edgecolor='black')
plt.title('Rating Numerator Distribution')
plt.ylabel('Frequency')
plt.xlabel('Rating_Numerator')
```

```python
df['rating_numerator'].value_counts()
```

```python
plt.hist(df['rating_denominator'].value_counts(), bins= 15, color='green', rwidth=20,edgecolor='black')
plt.title('Rating Denominator Distribution')
plt.ylabel('Frequency')
plt.xlabel('Rating_Denominator')
```
