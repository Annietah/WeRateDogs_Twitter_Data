## Analysis Description
The cleaned WeRateDogs twitter data was used to come up with the following  analyses and visualizations. 


### 1. Most common dog stage
To analyze this, we need to plot a bar graph. The most common dog type is pupper with a total of 206 dogs, then followed by doggo with a total of 63 dogs. However, there are several dog stage names that were invalid (None values). Therefore, this result might be affected when one can obtain the actual dog stage for the none values. 


### 2. Top 5 dog breeds by total retweet count
It is essential to know the most common dog breed from the tweets. We can also visualize this using  a bar chart. From the analysis, the most popular dog breed is Gold_retriever. This breed has the most number of retweet counts. It is followed by Labrador_retriever, Pembroke, Samoyed, and Oihuahua.



### 3. Relationship between favorite_count and retweet_count
       
To analyse the relationship between favorite count and retweet count, we need to plot a scatter plot. From the scatter plot, it is clear that there is a very positive linear relationship between number of favorite count and number of retweet count for a given tweet. This is true since, in twitter, it is common for favorite tweets to receive most retweets.

### 4. Ratings Distribution
We need to know how the dogs were rated on twitter. This analysis can be done by determining the value counts for each ratings. The most popular rating for the numerator is 12.0 while the most popular rating for the denominator is 10.0
